```python
ig = {'ft': [100,90,80,70,60,50,40,30,20,10],
        'a': [90.7,90.5,90.6,90.9,89.6,70.9,70.3,70.8,63.8,62.9]
       } 
ig_df = pd.DataFrame(ig,columns=['ft','a']) 

cr = {'ft': [100,90,80,70,60,50,40,30,20,10],
        'a': [79.8,79.9,78.6,77.5,77.1,76.5,74.4,70.6,66.3,58.3]
       } 
cr_df = pd.DataFrame(cr,columns=['ft','a'])

rff = {'ft': [100,90,80,70,60,50,40,30,20,10],
        'a': [92.9,93.2,93.6,93.0,92.7,91.8,92.2,91.2,89.9,87.4]
       } 
rff_df = pd.DataFrame(rff,columns=['ft','a'])

gr = {'ft': [100,90,80,70,60,50,40,30,20,10],
        'a': [77.0,75.4,70.6,71.7,64.9,61.8,59.4,58.2,55.5,45.6]
       } 
gr_df = pd.DataFrame(gr,columns=['ft','a'])

chi = {'ft': [100,90,80,70,60,50,40,30,20,10],
        'a': [92.5,92.7,92.7,92.8,89.6,89.0,89.6,70.0,68.9,68.5]
       } 
chi_df = pd.DataFrame(chi,columns=['ft','a'])
```

```python
plt.figure(figsize=(10, 7))
plt.plot(ig_df['ft'], ig_df['a'], c='red', label='InfoGain')
plt.plot(cr_df['ft'], cr_df['a'], c='green', label='Correlation')
plt.plot(rff_df['ft'], rff_df['a'], c='blue', label='Relief')
plt.plot(gr_df['ft'], gr_df['a'], c='orange', label='Gain Ratio')
plt.plot(chi_df['ft'], chi_df['a'], c='yellow', label='Chi Square')

plt.title('Decision Tree Accuracy Against Features Graph')
plt.xlabel('Number Of Features')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```
