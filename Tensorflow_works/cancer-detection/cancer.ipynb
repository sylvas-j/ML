```python
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.head()
```

## LR

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# divide the dataset into train and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
df.shape,X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
X_train
```

#### Training

```python
# apply Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

#### Predictions

```python
# implemented our model through logistic regression
y_pred = lr.predict(X_test)
y_pred
```

```python
lr_acc = accuracy_score(y_test, y_pred)
print(lr_acc)
```

```python
tempResults = pd.DataFrame({'Algorithm':['Logistic Regression Method'], 'Accuracy':[lr_acc]})
tempResults
```

#### Discussion

```python
result = pd.DataFrame({'pred':y_pred, 'test':y_test})
result
```

```python
result['actual_pred'] = result['pred'].map({1: 'M', 0: 'B'})
result['actual_test'] = result['test'].map({1: 'M', 0: 'B'})
result.head()
```

```python
sns.countplot(result['actual_pred'], label="Count",)
sns.countplot(result['actual_test'], label="Count",)
```
