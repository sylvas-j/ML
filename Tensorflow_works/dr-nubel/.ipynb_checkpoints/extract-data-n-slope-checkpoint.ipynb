```python
def extract_sheets_values(wb):
    for i in range(len(wb.sheetnames)):
        sheet = wb[wb.sheetnames[i]]
        print(wb.sheetnames[i])
        df = pd.DataFrame(sheet.values)
        if type(df.iloc[1][1]) == str:
            dff = df.drop([0],axis=1).drop([0,1],axis=0).rename(columns={1:'strain',2:'stress'}, inplace=False)
        else:
            dff = df.drop([0],axis=1).drop([0],axis=0).rename(columns={1:'strain',2:'stress'}, inplace=False)
        dff.to_csv('../../../datasets/dr-nubel/strain_stress/{}.csv'.format(wb.sheetnames[i]),index=False)
    # dff.head(2)
```

```python
import matplotlib.pyplot as plt
import plotly.express as px
from dash import Dash, dcc, html, Input, Output
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
1%2
```

```python
shnames = wb.sheetnames
r = len(shnames)//2+1
c = 2

g_r = 1
g_c = 1
fig = make_subplots(rows=r, cols=c,subplot_titles=(shnames))
for i in range(len(wb.sheetnames)):
    df = pd.read_csv('../../../datasets/dr-nubel/strain_stress/{}.csv'.format(wb.sheetnames[i]))
    if i > 1:
        if i%2 == 0:
            g_r+=1
            g_c=1
    
    fig.add_trace(go.Scatter(x=df.strain, y=df.stress),row=g_r, col=g_c)
    # print(i,g_r,g_c)
    g_c += 1
    
fig.update_layout(height=5000, width=800,title_text="Subplots of the different data")
# fig.update_yaxes(automargin=True)
fig.update_layout(margin=dict(l=30, r=20, t=40, b=20),paper_bgcolor="LightSteelBlue")
fig.show()
```

```python
# fig = plt.figure(figsize=(5,10))
fig = make_subplots(rows=1, cols=2)

fig = px.line(data_frame=df,x='strain' ,y ='stress', labels={'x':'Strain', 'y':'Stress'},width=400, height=400)
fig.update_layout(title_text=wb.sheetnames[0])
fig.update_layout(margin=dict(l=20, r=20, t=30, b=20),paper_bgcolor="LightSteelBlue")
# fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")

fig.show()
```
