```python

def getByTemp(df,t):
    aaZeroT = df.loc[df['temp'] == t]
    aaZeroNoT = aaZeroT.drop(['temp'], axis=1)
    return aaZeroNoT
```

```python
# titanic_df[titanic_df['Survived']==1]['Survived']
def hydrateNaclCurvePred2(df,x,y,y1,y2,c,title,xlab,ylab,cv1,cv2,cv3):
    
    legend_prop = {'weight':'bold','size':20}
    
    plt.plot(df[x].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             'gp-',linewidth=2, markersize=12, label =str(cv1)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             'gp:',linewidth=2, markersize=12, label =str(cv2)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             'gp--',linewidth=2, markersize=12, label =str(cv3)+'% '+str(c))
    
    plt.plot(df[x].where(df[c] == cv1).dropna(), df[y1].where(df[c] == cv1).dropna(),
             'bo-',linewidth=2, markersize=12, label =str(cv1)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv2).dropna(), df[y1].where(df[c] == cv2).dropna(),
             'bo:',linewidth=2, markersize=12, label =str(cv2)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv3).dropna(), df[y1].where(df[c] == cv3).dropna(),
             'bo--',linewidth=2, markersize=12, label =str(cv3)+'% '+str(c))
    
    plt.plot(df[x].where(df[c] == cv1).dropna(), df[y2].where(df[c] == cv1).dropna(),
             'mp-',linewidth=2, markersize=12, label =str(cv1)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv2).dropna(), df[y2].where(df[c] == cv2).dropna(),
             'mp:',linewidth=2, markersize=12, label =str(cv2)+'% '+str(c))
    plt.plot(df[x].where(df[c] == cv3).dropna(), df[y2].where(df[c] == cv3).dropna(),
             'mp--',linewidth=2, markersize=12, label =str(cv3)+'% '+str(c))
    
    plt.title(title, fontweight='bold',fontsize=15)
    plt.xlabel(xlab,fontsize=15)
    plt.ylabel(ylab,fontsize=15)
    plt.legend(prop=legend_prop)    
   


#  titanic_df[titanic_df['Survived']==1]['Survived']
def hydrateNaclCurvePred3(df,x,x1,x2,y,c,title,xlab,ylab,cv1,cv2,cv3):
    
    
    plt.plot(df[x].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             'gp-',linewidth=2, markersize=15, label =str(cv1)+'wt%'+str('nacl')+ ' : EXP')
    plt.plot(df[x].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             'gp:',linewidth=2, markersize=15,label =str(cv1)+'wt%'+str('nacl')+ ' : EXP')
    plt.plot(df[x].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             'gp--',linewidth=2, markersize=15, label =str(cv1)+'wt%'+str('nacl')+ ' : EXP')

    plt.plot(df[x1].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             'ms-',linewidth=2, markersize=10, label =str(cv2)+'wt%'+str('nacl')+ ' : ANN')
    plt.plot(df[x1].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             'ms:',linewidth=2, markersize=10,label =str(cv2)+'wt%'+str('nacl')+ ' : ANN')
    plt.plot(df[x1].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             'ms--',linewidth=2, markersize=10,label =str(cv2)+'wt%'+str('nacl')+ ' : ANN')    

    plt.plot(df[x2].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             'bo-',linewidth=2, markersize=6, label =str(cv3)+'wt%'+str('nacl')+ ' : CBANN')
    plt.plot(df[x2].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             'bo:',linewidth=2, markersize=6,label =str(cv3)+'wt%'+str('nacl')+ ' : CBANN')
    plt.plot(df[x2].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             'bo--',linewidth=2, markersize=6,label =str(cv3)+'wt%'+str('nacl')+ ' : CBANN')
    

    legend_prop = {'weight':'bold','size':12}
    
    plt.title(title, fontweight='bold',fontsize=12)
    plt.xlabel(xlab,fontsize=15)
    plt.ylabel(ylab,fontsize=15)
    plt.legend(prop=legend_prop)    
   
```

```python
def hydrateScatterPlot(df,x,x1,x2,y,c,title,xlab,ylab,cv1,cv2,cv3):
    
    plt.scatter(df[x].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             c='red', s=50, label ='EXP')
    plt.scatter(df[x1].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             c='green', s=50, marker='s', label ='ANN')
    plt.scatter(df[x2].where(df[c] == cv1).dropna(), df[y].where(df[c] == cv1).dropna(),
             c='blue', s=50, marker='p', label ='CBANN')

    
    plt.scatter(df[x].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             c='red', s=50)
    plt.scatter(df[x1].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             c='green', s=50, marker='s')
    plt.scatter(df[x2].where(df[c] == cv2).dropna(), df[y].where(df[c] == cv2).dropna(),
             c='blue', s=50, marker='p')

    
    plt.scatter(df[x].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             c='red', s=50)
    plt.scatter(df[x1].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             c='green', s=50, marker='s')
    plt.scatter(df[x2].where(df[c] == cv3).dropna(), df[y].where(df[c] == cv3).dropna(),
             c='blue', s=50, marker='p')

    legend_prop = {'weight':'bold','size':10}
    
    plt.title(title, fontweight='bold',fontsize=20)
    plt.xlabel(xlab,fontsize=15)
    plt.ylabel(ylab,fontsize=15)
#     plt.legend(bbox_to_anchor=(0.5, 0.7), ncol=1,fancybox=True, loc='upper center',prop=legend_prop)
    plt.legend(bbox_to_anchor=(0.3, 0.9), ncol=1,fancybox=True, loc='right',prop=legend_prop)
```

```python
# FOR AA-NACL DATASET
plt.figure(figsize=(20, 25))
# plt.rcParams["font.weight"] = "bold"
# plt.rcParams["axes.labelweight"] = "bold"
j = 1
tempt = ['0','4','15','25']
vert = ['aa','aa1','aa2']
dataset = [aaZeroNoT,aaNoT_4,aaNoT_15,aaNoT_25]
for i in range(4):
    plt.subplot(4,2,j)
    hydrateScatterPlot(dataset[i],'nacl','nacl1','nacl2','velocity','aa',tempt[i]+'\u00B0C'
                 ,'NaCl Concentration','Velocity',1,1.5,2)
    j = j + 1

# FOR CONDUCTIVITY
j = 5
for i in range(4):
    plt.subplot(4,2,j)
    hydrateScatterPlot(dataset[i],'nacl','nacl1','nacl2','conductivity','aa',tempt[i]+'\u00B0C'
                 ,'NaCl Concentration','Conductivity',1,1.5,2)
    j = j + 1
```
